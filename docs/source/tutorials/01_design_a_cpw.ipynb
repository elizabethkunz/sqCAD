# Tutorial 1: Designing a CPW Element

This essentially acts exactly like TX Line.

You will need the following values:
- Width (11.7um)
- Spacing (5.1um)
- Thickness (200nm)
- Substrate Height (525um)
- Relative Permittivity Constant (11.45 for ultracold silicon)
- Resistivity (set it as low as possible)

Choose a range of frequencies that you are interested in. It does not really matter for now. 

Now, Use the CPW element with the values input and run it:

```python
from skrf import Frequency
from skrf.media import CPW, Coaxial, Media

freq = Frequency(5, 10, 10001, 'GHz') 



cpw =  CPW(freq, w = 11.7e-6,
            s = 5.1e-6, #spacing = 5.1um
            t=0.2e-6, #thickness = 200nm
            h=525e-6, #substrate height = 525um
            rho=1e-19, #closest to 0 this thing goes
            ep_r = 11.45, #ultracold silicon
            has_metal_backside=True,
            tand=0,
            )
cpw
```

Now define a line that satisfies 180 degrees (lambda/2), or 90 degrees for a lambda/4 resonator:

```python
cpw.line(d=180,unit='deg', name='line')
```

To get the electrical length at a given frequency, choose a single frequency like so:

```python
freq = Frequency(7.8, 7.8, 1, 'GHz')    
cpw =  CPW(freq, w = 11.7e-6,
            s = 5.1e-6, #spacing = 5.1um
            t=0.2e-6, #thickness = 200nm
            h=525e-6, #substrate height = 525um
            rho=1e-19, #closest to 0 this thing goes
            ep_r = 11.45, #ultracold silicon
            has_metal_backside=True,
            tand=0,
            )

length = 8e-3
cpw.electrical_length(length, deg=True)
```
